```python
#data cleaning with data frames
example2 = pd.DataFrame({
     'column A': [1, 2, np.nan],
     'column B': [np.nan, 2, 3],
     'column C': [1, 'a', 3],
     'column D': [1, np.nan, np.nan]
    })

example2.info()
```

```python
example2
```

```python
example2.dropna(how='any')
```

```python
example2.dropna(how='all')
```

```python
example2.fillna({col: example2[col].mean() for col in example2.columns if example2[col].dtype == 'float64'})
```

```python
example2.fillna(method='ffill') #forward fill           
```
